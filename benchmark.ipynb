```python
import yfinance as yf
import pandas as pd
import numpy as np

def download_and_process_benchmark(start_date="1987-01-01", end_date="2021-12-31"):
    sp500 = yf.download("^GSPC", start=start_date, end=end_date, interval="1mo", auto_adjust=True)
    sp500.index = sp500.index.to_period('M').to_timestamp('M')
    sp500['SP500_Return'] = sp500['Close'].pct_change()
    
    tbill = yf.download("^IRX", start=start_date, end=end_date, interval="1mo", auto_adjust=True)
    tbill.index = tbill.index.to_period('M').to_timestamp('M')
    tbill['Rf_Monthly'] = (1 + tbill['Close']/100) ** (1/12) - 1
    
    benchmark = pd.concat([sp500['SP500_Return'], tbill['Rf_Monthly']], axis=1)
    benchmark.columns = ['SP500_Return', 'Rf_Monthly']
    benchmark = benchmark.dropna()
    benchmark['Excess_Return'] = benchmark['SP500_Return'] - benchmark['Rf_Monthly']
    
    return benchmark

benchmark = download_and_process_benchmark()
benchmark.to_csv("SP500_Excess_Return_1987_2021.csv")
```
